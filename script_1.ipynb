```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
r = requests.get("https://en.wikipedia.org/wiki/Toy_Story_3")

# Convert to a Beautiful Soup Obejct
soup = bs(r.content)
contents = soup.prettify()
```

```python
print(contents)
```

```python
info_box = soup.find(class_= "infobox vevent")
print(info_box.prettify())
```

```python
# How to get all table rows
info_rows = info_box.find_all("tr")
for row in info_rows:
    print(row.prettify())
```

```python
# Making a dictionary and adding data from the table
movie_info = {}


# Defining a function to handle list in data
# this is for when content_value is a list type
def get_content_value(row_data):
    if row.find("li"):
        return [li.get_text().replace("\xa0", " ") for li in row_data.find_all("li")]
    else:
        return row_data.get_text().replace("\xa0"," ") # removing xa0 from text 


for index, row in enumerate(info_rows):
    if index == 0:
        movie_info['title'] = row.find("th").get_text()
    elif index == 1:
        continue
    else:
        content_key = row.find("th").get_text()
        content_value = get_content_value(row.find("td"))
        movie_info[content_key] = content_value 
        



         
```
